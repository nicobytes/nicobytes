## Challenge: Reverse a linked list (linked list in, string out)

### Problem

You are given the **head** of a **non-empty or empty** singly linked list (`ListNode` chain). The first node is the **head**; the last node’s `next` is `None`.

Your task:

1. **Reverse** the linked list in place or by rewiring pointers (classic linked-list reversal—not a trick based on printing backwards).
2. Return the reversed sequence as a single **`str`**, with values separated by **commas** (`,`), **head → tail** of the **reversed** list (no trailing comma).

Parsing a comma-separated string into nodes is **out of scope** for the main API: callers may build the list however they want; your function starts from `head`.

### I/O contract

| | |
|---|---|
| **Input** | `Optional[ListNode]` — head of the list; `None` means empty list |
| **Output** | `str` — comma-separated values after reversal, e.g. `"3,2,1"` |

### Requirements

- **`input`** is already a linked list (or `None`). **`output`** is a string.
- Use pointer manipulation on `ListNode` to reverse the structure.

### Edge cases to consider

- `head is None` → return `""` (or your documented convention for empty).
- Single node → output is that single `val` with no commas.
- Formatting: commas only between values, never a trailing comma.

### Examples

If the list is `1 → 2 → 3 → None`, after reversal your string should be `"3,2,1"`.

If the list is `hello → world → None`, output `"world,hello"`.

Implement **`reverse_linked_list_to_string`** in the next cell.

```python
from __future__ import annotations

from typing import Optional


class ListNode:
    """Singly linked list node."""

    def __init__(self, val: str, next: Optional["ListNode"] = None) -> None:
        self.val = val
        self.next = next


def reverse_linked_list_to_string(head: Optional[ListNode]) -> str:
    current = head 
    prev = None
    while current:
        temp = current.next 
        current.next = prev
        prev = current
        current = temp

    head = prev
    result = []
    while head:
        result.append(head.val)
        head = head.next
    return ",".join(result)


# --- Optional: manual checks (build lists by hand; uncomment when ready) ---
head = ListNode("1", ListNode("2", ListNode("3")))
assert reverse_linked_list_to_string(head) == "3,2,1"
assert reverse_linked_list_to_string(None) == ""
solo = ListNode("solo")
assert reverse_linked_list_to_string(solo) == "solo"
```
